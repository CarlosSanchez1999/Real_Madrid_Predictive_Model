```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
import pandas as pd
```

```python
# Load the CSV file
third_data = pd.read_csv('C:/Users/PC/Desktop/Projects_Repository/Projects/Real_Madrid_Predictive_Model/Second Data Cleaning Process/cleaned_data2.csv')
third_data. head()
```

```python
# Step 1: Convert the Date column to datetime type
third_data['Date'] = pd.to_datetime(third_data['Date'])
```

```python
# Step 2: Encode the "Result" column with Label Encoding
le = LabelEncoder()
third_data['Result'] = le.fit_transform(third_data['Result'])
```

```python
# Step 3: One-Hot Encoding for the home and away teams
third_data = pd.get_dummies(third_data, columns=['Home_Team', 'Away_Team'], drop_first=True)
```

```python
# Step 4: Scale the numerical columns
scaler = StandardScaler()
numerical_columns = [
    'Home_Goals', 'Away_Goals', 'Home_Shots', 'Away_Shots', 
    'Home_Shots_on_Target', 'Away_Shots_on_Target', 'Home_Corners', 
    'Away_Corners', 'Home_Yellow_Cards', 'Away_Yellow_Cards', 
    'Home_Red_Cards', 'Away_Red_Cards'
]
third_data[numerical_columns] = scaler.fit_transform(third_data[numerical_columns])
```
