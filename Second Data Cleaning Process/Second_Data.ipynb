```python
import pandas as pd

# Cargar el archivo CSV
df_cleaned = pd.read_csv('C:/Users/PC/Desktop/Projects_Repository/Projects/Real_Madrid_Predictive_Model/First Data Cleaning Process/cleaned_matches_all_seasons.csv')

# Ver las primeras filas del DataFrame para confirmar la carga
df_cleaned.head()
```

```python
#Fix dates and make them Int64
# Function to parse date with known formats
def parse_date(date_str):
    try:
        # Try common formats
        return pd.to_datetime(date_str, format="%d/%m/%y", dayfirst=True)
    except (ValueError, TypeError):
        try:
            return pd.to_datetime(date_str, format="%d-%m-%Y", dayfirst=True)
        except (ValueError, TypeError):
            return pd.NaT  # Return NaT if parsing fails

# Apply the parsing function
df_cleaned['Date'] = df_cleaned['Date'].apply(parse_date)

# Drop rows with unparseable dates
df_cleaned.dropna(subset=['Date'], inplace=True)

# Sort data by date to ensure chronological order
df_cleaned.sort_values('Date', inplace=True)

# Check the first few rows to confirm successful conversion and sorting
df_cleaned[['Date']].head()
```

```python
df_cleaned.to_csv('cleaned_data2.csv', index=False)
print("se exporto un nuevo csv en la carpeta actual")
```
